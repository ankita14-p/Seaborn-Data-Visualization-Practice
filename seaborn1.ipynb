#Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Note: Seaborn has some built-in dataset

```python
#total bill vs tip dataset
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#setting the default theme
sns.set_theme()
```

```python
#visuallize the data
sns.relplot(x="total_bill",y="tip",hue="smoker",style="smoker",size="size",col="time",data=tips)
```

```python
#loading the titanic dataset
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
sns.countplot(x="class",hue="who",data=titanic)
```

```python
sns.countplot(x="survived",hue="class",data=titanic)
```

#Bar Chart

```python
#house price dataset
from sklearn.datasets import fetch_california_housing
```

```python
california=fetch_california_housing()
print(california)
```

```python
#converting the np array of the california dataset to pandas dataframe
california_df=pd.DataFrame(california.data,columns=california.feature_names)
california_df.head()
```

#Distribution Plot

```python
california_df["Price"]=california.target
california_df.head()
```

```python
sns.displot(california_df["Price"])
```

```python
sns.distplot(california_df["Price"])
```

#Correlation Matrix

```python
correlation=california_df.corr()
california_df.shape
```

```python
#HeatMap
plt.figure(figsize=(9,9))
sns.heatmap(correlation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={"size":9},cmap="Reds")
```
